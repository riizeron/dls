```python
import numpy as np
import pandas as pd
import scipy.linalg as sla
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

### My custom regression

```python
class MyLinearRegression:
    def __init__(self, fit_intercept=True):
        self.fit_intercept = fit_intercept
    
    def fit(self, X, y):
        n, k = X.shape
        
        X_train = X
        
        if self.fit_intercept:
            X_train = np.hstack((X, np.ones((n, 1))))
            
        self.w = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y
        
        return self
    
    def predict(self, X):
        n, k = X.shape
        
        if self.fit_intercept:
            X_train = np.hstack((X, np.ones((n, 1))))
            
        y_pred = X_train @ self.w
        
        return y_pred
    
    def get_weights(self):
        return self.w
```

```python
from sklearn.model_selection import train_test_split
```

```python
def linear_expression(x):
    return 5 * x + 6
```

```python
objects_num = 50

X = np.linspace(-5, 5, objects_num)
y = linear_expression(X) + np.random.randn(objects_num) * 5

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5)
```

### Gradient optimisation

```python
 class MyGradientLinearRegressor(MyLinearRegression):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = None
        
    def fit(self, X, y, lr=0.01, max_iter=100):
        n, k = X.shape

        X_train = np.hstack((X, np.ones((n, 1)))) if self.fit_intercept else X

        if self.w is None:
            self.w = np.random.randn(k + 1 if self.fit_intercept else k)

        self.losses = []

        for iter_num in range(max_iter):
            y_pred = self.predict(X)
            self.losses.append(mean_squared_error(y_pred, y))

            grad = self._calc_gradient(X_train, y, y_pred)

            assert grad.shape == self.w.shape, f"gradient shape {grad.shape} is not equal weight shape {self.w.shape}"

            self.w -= lr * grad
            
        return self
    
    
    def _calc_gradient(self, X, y, y_pred):
        grad = (2 * (y_pred - y)[:, np.newaxis] * X).mean(axis=0)
        
        return grad
    
    def get_losses(self):
        return self.losses    
            
```

```python
regressor = MyGradientLinearRegressor() 

l = regressor.fit(X_train[:, np.newaxis], y_train).get_losses()

predicts = regressor.predict(X_test[:, np.newaxis])
w = regressor.get_weights()
w
```

### Stohastic gradient decent

```python
class MySGDLinearRegressor(MyGradientLinearRegressor):
    def __init__(self, n_samples=10, **kwargs):
        super().__init__(**kwargs)
        self.w = None
        self.n_samples = n_samples
        
    def _calc_gradient(self, X, y, y_pred):
        inds = np.random.choice(np.arange(X.shape[0]), size=self.n_samples, replace=False)
        
        grad = (2 * (y_pred[inds] - y[inds])[:, np.newaxis] * X[inds]).mean(axis=0)
        
        return grad
```

```python
regressor = MySGDLinearRegressor()

l = regressor.fit(X_train[:, np.newaxis], y_train).get_losses()

predicts = regressor.predict(X_test[:, np.newaxis])
regressor.get_weights()
```

### Logistic Regression

```python
def logit(x, w):
    return np.dot(x, w)

def sigmoid(h):
    return 1. / (1 + np.exp(-h))

class MyLogisticRegressor:
    def __init__(self):
        self.w = None
        
    def fit(self, X, y, max_iter=100, lr=0.1):
        n, k = X.shape
        
        if self.w is None:
            self.w = np.random.randn(k + 1)
            
        X_train = np.concatenate((np.ones((n,1)), X), axis=1)
        
        self.losses = []
        
        for iter_num in range(max_iter):
            z = sigmoid(logit(X_train, self.w))
            
            grad = np.dot(X_train.T, (z - y)).mean(axis=0)
            
            self.w -= lr * grad
            
            self.losses.append(self._loss(y, z))
            
        return self
    
    def predict_proba(self, X):
        n, k = X.shape
        X_ = np.concatenate((np.ones((n, 1)), X), axis=1)
        return sigmoid(logit(X_, self.w))
    
    def predict(self, X, threshold=0.5):
        return self.predict_proba(X) >= threshold
    
    def get_weights(self):
        return self.w
    
    def get_losses(self):
        return self.losses
    
    def _loss(self, y, p):
        p = np.clip(p, 1e-10, 1 - 1e-10)
        return np.mean(y * np.log(p) + (1 - y) * np.log(1 - p))
```
